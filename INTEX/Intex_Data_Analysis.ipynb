## Homelessness by Age Demographics in Utah

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV data into a DataFrame
df = pd.read_csv('Simplified_PIT_data.csv')

# Filter the DataFrame to include only rows where the CoC Name starts with "UT"
df_utah = df[df["CoC Number"].str[:2] == "UT"]

# Select the relevant columns for the bar chart
df_utah = df_utah[['CoC Name', 'Sheltered Total Homeless - Under 18', 'Sheltered Total Homeless - Age 18 to 24', 'Sheltered Total Homeless - Over 24']]

# Rename columns for better readability
df_utah.columns = ['CoC Name', 'Under 18', 'Age 18 to 24', 'Over 24']

# Set the index to CoC Name for better plotting
df_utah.set_index('CoC Name', inplace=True)

# Define a custom color palette
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Plotting the bar chart with custom colors and spacing
ax = df_utah.plot(kind='bar', figsize=(12, 8), color=colors, width=0.8)

# Add titles and labels
ax.set_title('Homelessness by Age Demographics in Utah', fontsize=14)
ax.set_xlabel('CoC Name', fontsize=12)
ax.set_ylabel('Number of Homeless Individuals', fontsize=12)

# Customize the tick parameters and rotation
plt.xticks(rotation=45, ha='right')

# Add a grid for better readability
plt.grid(axis='y', linestyle='--', linewidth=0.7)

# Add a legend
plt.legend(title='Age Group', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()

```

## Percentage of Homelessness by Race

```python
import pandas as pd
import matplotlib.pyplot as plt


# List of race columns
race_columns = [
    'Sheltered Total Homeless - White',
    'Sheltered Total Homeless - Black, African American, or African',
    'Sheltered Total Homeless - Asian or Asian American',
    'Sheltered Total Homeless - American Indian, Alaska Native, or Indigenous',
    'Sheltered Total Homeless - Native Hawaiian or Other Pacific Islander',
    'Sheltered Total Homeless - Multiple Races'
]

# Calculate the total homelessness for each race
total_homeless_by_race = df[race_columns].sum()

# Calculate the total homelessness
total_homeless = total_homeless_by_race.sum()

# Calculate the percentage of homelessness by race
percentage_by_race = (total_homeless_by_race / total_homeless) * 100

# Plotting the bar chart
plt.figure(figsize=(10, 6))
percentage_by_race.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b'])
plt.title('Percentage of Homelessness by Race')
plt.xlabel('Race')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```

## Significane of Age Group & Sheltered or Unsheltered 

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Select relevant columns for age groups of unsheltered homeless
age_columns_unsheltered = [
    'Unsheltered Homeless - Under 18',
    'Unsheltered Homeless - Age 18 to 24',
    'Unsheltered Homeless - Over 24'
]

# Ensure the data is numeric and drop missing values
unsheltered_under_18 = pd.to_numeric(df['Unsheltered Homeless - Under 18'], errors='coerce').dropna()
unsheltered_age_18_to_24 = pd.to_numeric(df['Unsheltered Homeless - Age 18 to 24'], errors='coerce').dropna()
unsheltered_over_24 = pd.to_numeric(df['Unsheltered Homeless - Over 24'], errors='coerce').dropna()

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(unsheltered_under_18, unsheltered_age_18_to_24, unsheltered_over_24)

# Round the p-value to 2 decimal places
p_value_rounded = round(p_value, 2)

# Print the results
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value_rounded}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value_rounded < alpha:
    print("The differences between age groups are statistically significant.")
else:
    print("The differences between age groups are not statistically significant.")
```

## Significance of Sheltered & Unsheltered 

```python
import pandas as pd
from scipy import stats

# Load your CSV data into a DataFrame
df = pd.read_csv('Simplified_PIT_data.csv')

# Ensure the data is numeric and handle missing values
df['Unsheltered Homeless - Under 18'] = pd.to_numeric(df['Unsheltered Homeless - Under 18'], errors='coerce').fillna(0)
df['Unsheltered Homeless - Age 18 to 24'] = pd.to_numeric(df['Unsheltered Homeless - Age 18 to 24'], errors='coerce').fillna(0)
df['Unsheltered Homeless - Over 24'] = pd.to_numeric(df['Unsheltered Homeless - Over 24'], errors='coerce').fillna(0)
df['Sheltered Total Homeless - Under 18'] = pd.to_numeric(df['Sheltered Total Homeless - Under 18'], errors='coerce').fillna(0)
df['Sheltered Total Homeless - Age 18 to 24'] = pd.to_numeric(df['Sheltered Total Homeless - Age 18 to 24'], errors='coerce').fillna(0)
df['Sheltered Total Homeless - Over 24'] = pd.to_numeric(df['Sheltered Total Homeless - Over 24'], errors='coerce').fillna(0)

# Sum the values for each age group to create a total for unsheltered and sheltered
total_Unsheltered = df['Unsheltered Homeless - Under 18'] + df['Unsheltered Homeless - Age 18 to 24'] + df['Unsheltered Homeless - Over 24']
total_Sheltered = df['Sheltered Total Homeless - Under 18'] + df['Sheltered Total Homeless - Age 18 to 24'] + df['Sheltered Total Homeless - Over 24']

# Perform T-Test to compare the two groups
t_stat, p_value = stats.ttest_ind(total_Unsheltered, total_Sheltered)

# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value:.2f}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("The difference between sheltered and unsheltered populations is statistically significant.")
else:
    print("The difference between sheltered and unsheltered populations is not statistically significant.")

```
